```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file = "Purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data = pd.DataFrame(purchase_data)
```

## Player Count

* Display the total number of players


```python
# Display the first 4 rows with Coloumn Names
purchase_data.head()
```

```python
#Identify the unique players within the Data set
players = purchase_data["SN"].unique()

#Count the number of unique players
total_players = len(players)

#Print the total number of players
print("The total number of players is " + str(total_players))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate Number of Unique Items
unique_items = purchase_data["Item ID"].unique()
number_of_unique_items = len(unique_items)
number_of_unique_items

# Calculate Average Price
price = purchase_data["Price"].mean()
average_price = round(price,2)
average_price

# Calculate Total Purchases
total_purchases = purchase_data["Purchase ID"].count()
total_purchases

# Calculate Total Revenue
total_revenue = purchase_data["Price"].sum()
total_revenue

# Formating Average Price & Total Revenue
summary_table_df = pd.DataFrame({"Number of Unique Items" :[number_of_unique_items],"Average Price":[average_price],"Number of Purchases": [total_purchases], "Total Revenue" : [total_revenue]})

# Add $ to Average Price and Total Revenue

summary_table_df["Average Price"] = summary_table_df["Average Price"].map("${:.2f}".format)
summary_table_df["Total Revenue"] = summary_table_df["Total Revenue"].map("${:.2f}".format)

# Display Data Frame
summary_table_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Group Data by Gender
grouped_purchase_data = purchase_data.groupby(["Gender"])
print(grouped_purchase_data)

# Count Gender Total
gender = grouped_purchase_data["SN"].nunique()
gender
total_gender = gender.sum()
total_gender

# Calculate Gender Percentage
percentage = (gender/total_gender)*100
percentage

# Enter values into Data Frame
gender_demographics_df = pd.DataFrame({"Total Count" : gender, "Percentage of Players" : percentage})

# Add % to Percentage of Players
gender_demographics_df['Percentage of Players'] = gender_demographics_df["Percentage of Players"].map("{:.2f}%".format)

# Display Data Frame
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group Data by Gender
grouped_purchase_data = purchase_data.groupby(["Gender"])
print(grouped_purchase_data)

# Count the total number of Purchase ID per Gender
purchase_count = grouped_purchase_data["Purchase ID"].count()
purchase_count

# Calcuate Average Purchase Price
average_price = round(grouped_purchase_data["Price"].mean(),2)
average_price

# Calcuate Total Purchase Value
total_purchase = round(purchase_count*average_price,2)
total_purchase

# Count the unique number of players
unique_number_of_players = grouped_purchase_data["SN"].nunique()
unique_number_of_players

# Calculate the average total purchase per person
average_purchase_price = round(total_purchase/unique_number_of_players,2)
average_purchase_price

# Create Data Frame
purchasing_analysis_df = pd.DataFrame({"Purchase Count" : purchase_count, "Average Purchase Price" : average_price, "Total Purchase Value" : total_purchase, "Avg Total Purchase per Person" : average_purchase_price})

# Add $ to Average Purchase Price, Total Purchase Value, Average Total Purchase Per Person
purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_df["Avg Total Purchase per Person"] = purchasing_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

# Display Data Frame
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create Bins for age group
bins = [0 ,9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 50.99]
age_group = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Cut data in age group bins
purchase_data["Age Range"]= pd.cut(purchase_data["Age"], bins, labels=age_group)
purchase_data.head()

# Group Data by Age Range
age_count = purchase_data.groupby('Age Range')['SN'].nunique()
age_count

# Calculate Total Count
total_count = age_count.sum()
total_count

# Calculate Percentage of Players
percentage = ((age_count/total_count)*100)
percentage

# Create Data frame
age_demographics_df = pd.DataFrame({"Total Count" : age_count, "Percentage of Players" : percentage})

# Add % to Percentage of Players
age_demographics_df['Percentage of Players'] = age_demographics_df["Percentage of Players"].map("{:.2f}%".format)

# Display Data Frame
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create Bins for age group
bins = [0 ,9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 50.99]
age_group = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Cut data in age group bins
purchase_data["Age Range"]= pd.cut(purchase_data["Age"], bins, labels=age_group)
purchase_data.head()

# Group Data by Age Range
grouped_purchase_data = purchase_data.groupby(["Age Range"])

# Calculate Purchase Count
purchase_count = grouped_purchase_data["Purchase ID"].count()
purchase_count

# Calculate Age Count
age_count = grouped_purchase_data['SN'].nunique()
age_count

# Calculate Avergae Purchase Price
average_price = round(grouped_purchase_data["Price"].mean(),2)
average_price

# Calculate Total Purchase Value
total_value = purchase_count*average_price
total_value

# Calculate Avg Total Purchase Per Person
avg_total = round(total_value/age_count,2)
avg_total

# Create Data frame
purchasing_analysis_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price" : average_price, "Total Purchase Value" : total_value, "Avg Total Purchase Per Person" : avg_total})

# Add $ to Average Purchase Price, Total Purchase Value, Average Total Purchase Per Person
purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_df["Avg Total Purchase Per Person"] = purchasing_analysis_df["Avg Total Purchase Per Person"].map("${:.2f}".format)

# Display Data Frame
purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group Data by SN
grouped_purchase_data = purchase_data.groupby(["SN"])
grouped_purchase_data

# Calculate Purchase Count
purchase_count = grouped_purchase_data["Purchase ID"].count()
purchase_count

# Calculate Average Pruchase Price
average_price = round(grouped_purchase_data["Price"].mean(),2)
average_price

# Calculate Total Purchase Price
total_purchase_price = purchase_count*average_price
total_purchase_price

# Create Data frame
top_spenders_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_price, "Total Purchase Value": total_purchase_price})

# Sort Total Purchase Price Highest to Lowest
top_spenders_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False)

# Add $ to Average Purchase Price and Total Purchase Value
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)

# Display Data Frame
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group Data by Item ID & Item Name
grouped_purchased_data= purchase_data.groupby(["Item ID", "Item Name"])
grouped_purchased_data

# Calculate Purchase Count
purchase_count = grouped_purchased_data["Purchase ID"].count()
purchase_count

# Calculate Item Price
item_price = round(grouped_purchased_data["Price"].mean(),2)
item_price

# Calculate Total Price
total_price = purchase_count*item_price
total_price

# Create Data frame
most_popular_item_df = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": item_price, "Total Purchase Value" : total_price})

# Sort Purchase Count Highest to Lowest
most_popular_item_df = most_popular_item_df.sort_values("Purchase Count", ascending=False)

# Add $ to Item Price and Total Purchase Value
most_popular_item_df["Item Price"] = most_popular_item_df["Item Price"].map("${:.2f}".format)
most_popular_item_df["Total Purchase Value"] = most_popular_item_df["Total Purchase Value"].map("${:.2f}".format)

# Display Data Frame
most_popular_item_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Create Data frame
most_popular_item_df = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": item_price, "Total Purchase Value" : total_price})

# Sort Total Purchase Value Highest to Lowest
most_profitable_items_df = most_popular_item_df.sort_values("Total Purchase Value", ascending=False)

# Add $ to Item Price and Total Purchase Value
most_profitable_items_df["Item Price"] = most_popular_item_df["Item Price"].map("${:.2f}".format)
most_profitable_items_df["Total Purchase Value"] = most_popular_item_df["Total Purchase Value"].map("${:.2f}".format)

# Display Data Frame
most_profitable_items_df.head()
```
